```python
import pandas as pd
import numpy as np
import regex as re
```

```python
#Load dataset
df_council = pd.read_excel('LocalityFinder_Sep22.xlsx')
df_council.columns = df_council.iloc[1]
#Select row 2 as column name
df_council = df_council.drop(0)
df_council = df_council.drop(1)
#Select only needed column
df_council = df_council.iloc[:, 0:3]
#Rename column
df_council = df_council.rename(columns={'Locality Name': 'Suburb'})
df_council = df_council.rename(columns={'Post_x000D_\nCode': 'Postcode'})
df_council = df_council.rename(columns={'Municipality_x000D_\nName': 'Council'})
#Format council name
df_council['Council'] = df_council['Council'].str.replace('City', '')
df_council['Council'] = df_council['Council'].str.replace('Council', '')
df_council['Council'] = df_council['Council'].str.replace('Shire', '')
df_council['Council'] = df_council['Council'].str.rstrip()
```

```python
#Load dataset
df_Course_Provider = pd.read_excel('004-victorianvetcoursesdata.xlsx', sheet_name='Course_Provider')
df_Providers = pd.read_excel('004-victorianvetcoursesdata.xlsx', sheet_name='Providers')
df_Courses = pd.read_excel('004-victorianvetcoursesdata.xlsx', sheet_name='Courses')
df_Providers
```

```python
# select multiple columns by name
df_Courses = df_Courses[['ID', 'Course Title','Qualification Level','entry_requirements','description']]

#Remove Qualification Level from course title column
# define the regular expression pattern to match
pattern = r'Certificate\s+I\s+in\s+'
# use the `replace()` function to replace the matched pattern with an empty string
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

pattern = r'Certificate\s+II\s+in\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

pattern = r'Certificate\s+III\s+in\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

pattern = r'Certificate\s+IV\s+in\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

pattern = r'Advanced\s+Diploma\s+of\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

pattern = r'Diploma\s+of\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

pattern = r'Graduate\s+Certificate\s+in\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')


pattern = r'Graduate\s+Diploma\s+of\s+'
df_Courses['Course Title'] = df_Courses['Course Title'].str.replace(pattern, '')

df_Courses['Course Title'] = df_Courses['Course Title'].str.lstrip()
df_Courses['Course Title'] = df_Courses['Course Title'].str.rstrip()

df_Courses
```

```python
#Merged the dataset to get needed information
merged_course = pd.merge(df_Course_Provider, df_Courses, left_on='course_id', right_on='ID', how='left')
merged_course
```

```python
# select multiple columns by name
df_Providers = df_Providers[['ID', 'Provider Name','Site Name','Postcode','url','email']]
df_Providers
```

```python
df_council['Postcode'] = df_council['Postcode'].astype('int64')
#Filter the dataset to show needed council
filtered_Providers = df_Providers[df_Providers['Postcode'].isin(df_council['Postcode'])]
```

```python
course_location = pd.merge(filtered_Providers, df_council, left_on='Postcode', right_on='Postcode', how='left')

course_location = course_location.drop_duplicates(subset=['ID'])
course_location
```

```python
df_course_location = pd.merge(merged_course, course_location, left_on='campus_id', right_on='ID', how='left')
df_course_location = df_course_location[['Course Title','Qualification Level','Provider Name','Site Name','Council']]
df_course_location = df_course_location.dropna()
df_course_location = df_course_location.reset_index(drop=True)
df_course_location
```

```python
df_Courses_information = df_Courses[['Course Title','Qualification Level','entry_requirements','description']]
df_Courses_information = df_Courses_information.reset_index(drop=True)
df_Courses_information
```

```python
#Saving to CSV
df_course_location.to_csv("Courses_location.csv", index=False)
df_Courses_information.to_csv("Courses_information.csv", index=False)
```
