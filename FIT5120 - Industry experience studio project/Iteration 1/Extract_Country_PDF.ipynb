```python
import os
import tabula
import re
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Set the folder path that contains the PDF files
folder_path = "PDF Original Data/"
count = 0
# Loop through every file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".pdf") and count == 0:
        #Read file and Change country name to the same format
        base_df = tabula.read_pdf(folder_path+filename, pages='12')[0]
        country_name = re.findall(".+?(?=-Com)", filename)
        #Check filename
        print(country_name)
        #Remove unused column
        base_df = base_df.iloc[4:]
        base_df = base_df.iloc[:, 0:3]
        #Format the row
        base_df['Unnamed: 1'] = base_df['Unnamed: 1'].str.replace(",", "")
        base_df['Unnamed: 1'] = base_df['Unnamed: 1'].str.replace(" ", "")
        base_df = base_df.astype({'Unnamed: 1':'int'})
        base_df = base_df.astype({'Unnamed: 2':'float'})
        #Replace Country name
        country_name = country_name[0].replace('-',' ')
        #Rename column
        base_df = base_df.rename(columns={"Unnamed: 0": "Suburb", "Unnamed: 1": country_name,"Unnamed: 2": country_name + " by %"})
        base_df= base_df.reset_index()
        
        #Create two DF percentage and df_population
        percentage = base_df[['Suburb', country_name + " by %"]]
        df_pop = base_df.iloc[: , 1:3]      
        count =+1
    else :
        #Check filename
        print(filename)
        #This file have to be read at page 13
        if filename == "Israel-Community-Profile-2016-Census.pdf":
            df = tabula.read_pdf(folder_path+filename, pages='13')[0]
            country_name = re.findall(".+?(?=-Com)", filename)
            #Remove unused column
            df = df.iloc[4:]
            df = df.iloc[:, 0:3]
            #Format the row
            df['Unnamed: 1'] = df['Unnamed: 1'].str.replace(",", "")
            df['Unnamed: 1'] = df['Unnamed: 1'].str.replace(" ", "")
            df = df.astype({'Unnamed: 1':'int'})
            df = df.astype({'Unnamed: 2':'float'})
            #Replace Country name
            country_name = country_name[0].replace('-',' ')
            #Rename column
            df = df.rename(columns={"Unnamed: 0": "Suburb", "Unnamed: 1": country_name,"Unnamed: 2": country_name + " by %"}) 
            df = df.reset_index()
            
            
            #Merge two dataframe
            df_place = df[['Suburb', country_name + " by %"]]
            percentage = pd.concat((percentage, df_place['Suburb'].to_frame('Suburb'))).drop_duplicates('Suburb')
            percentage = percentage.merge(df_place, on='Suburb', how='left')
            
            #Merge two dataframe
            df = df.iloc[: , 1:3]
            df_pop = pd.concat((df_pop, df['Suburb'].to_frame('Suburb'))).drop_duplicates('Suburb')
            df_pop = df_pop.merge(df, on='Suburb', how='left')
            
        #This two countries have similar error
        elif filename == "Bosnia-Herzegovina-Community-Profile-2016-Census.pdf" or filename == "UAE-Community-Profile-2016-Census.pdf":
            df = tabula.read_pdf(folder_path+filename, pages='12')[0]
            country_name = re.findall(".+?(?=-Com)", filename)
            #Remove unused column
            df = df.iloc[4:]
            df = df.iloc[:, 0:3]
            #Format the row
            df['Unnamed: 1'] = df['Unnamed: 1'].str.replace(",", "")
            df['Unnamed: 1'] = df['Unnamed: 1'].str.replace(" ", "")
            df = df.astype({'Unnamed: 1':'int'})
            #Replace Country name
            country_name = country_name[0].replace('-',' ')
            #Rename column
            df = df.rename(columns={"Unnamed: 0": "Suburb", "Unnamed: 1": country_name})
            df = df.astype({country_name:'float'})
            df.columns.values[2] = country_name + " by %"
            
            df = df.reset_index()            
            df[country_name + " by %"] = df[country_name + " by %"].str.extract('(\d+.\d+)') 
            
            #Merge two dataframe
            df_place = df[['Suburb', country_name + " by %"]]
            percentage = pd.concat((percentage, df_place['Suburb'].to_frame('Suburb'))).drop_duplicates('Suburb')
            percentage = percentage.merge(df_place, on='Suburb', how='left')
            df = df.iloc[: , 1:3]
            df_pop = pd.concat((df_pop, df['Suburb'].to_frame('Suburb'))).drop_duplicates('Suburb')
            df_pop = df_pop.merge(df, on='Suburb', how='left')
        else:
            df = tabula.read_pdf(folder_path+filename, pages='12')[0]
            country_name = re.findall(".+?(?=-Com)", filename)
            #Remove unused column
            df = df.iloc[4:]
            df = df.iloc[:, 0:3]
            #Format the row
            df['Unnamed: 1'] = df['Unnamed: 1'].str.replace(",", "")
            df['Unnamed: 1'] = df['Unnamed: 1'].str.replace(" ", "")            
            df = df.astype({'Unnamed: 1':'int'})
            df = df.astype({'Unnamed: 2':'float'})
            country_name = country_name[0].replace('-',' ')
            #Rename column
            df = df.rename(columns={"Unnamed: 0": "Suburb", "Unnamed: 1": country_name,"Unnamed: 2": country_name + " by %"}) 
            df = df.reset_index()
            
            #Merge two dataframe
            df_place = df[['Suburb', country_name + " by %"]]
            percentage = pd.concat((percentage, df_place['Suburb'].to_frame('Suburb'))).drop_duplicates('Suburb')
            percentage = percentage.merge(df_place, on='Suburb', how='left')
            
            
            df = df.iloc[: , 1:3]
            df_pop = pd.concat((df_pop, df['Suburb'].to_frame('Suburb'))).drop_duplicates('Suburb')
            df_pop = df_pop.merge(df, on='Suburb', how='left')

```

```python
#Remove the other area and Victoria row to last row
other = df_pop.iloc[20].copy()
victoria = df_pop.iloc[21].copy()
df_pop.drop(index=20, inplace=True)
df_pop.drop(index=21, inplace=True)
df_pop = df_pop.append(other)
df_pop = df_pop.append(victoria)
# reset the index of the DataFrame
df_pop = df_pop.reset_index(drop=True)
```

```python
#Remove the other area and Victoria row to last row
other = percentage.iloc[20].copy()
victoria = percentage.iloc[21].copy()
percentage.drop(index=20, inplace=True)
percentage.drop(index=21, inplace=True)
percentage = percentage.append(other)
percentage = percentage.append(victoria)
# reset the index of the DataFrame
percentage = percentage.reset_index(drop=True)
```
