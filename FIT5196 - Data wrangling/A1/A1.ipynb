```python
dirty_df.head()
```

Create a function to calucate distance between two countries.

```python
##This function slightly adjusted from 
##https://stackoverflow.com/questions/29545704/fast-haversine-approximation-python-pandas/29546836#29546836 and 
##https://www.adamsmith.haus/python/answers/how-to-find-the-distance-between-two-lat-long-coordinates-in-python

def haversine(lat1, lon1, lat2, lon2):

    earth_radius = 6378.0
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2

    distance = earth_radius * 2 * np.arcsin(np.sqrt(a))
    
    return np.around(distance,2)

dirty_df['dist'] =  haversine(dirty_df['latitude_x'], dirty_df['longitude_x'],dirty_df['latitude_y'], dirty_df['longitude_y'])
```

## 4. Finding the outlier

Calculate residuals of the linear model.

```python
a_outlier = sm.OLS(adonis_ol_df[y], adonis_ol_df[x].assign(const=1))
c_outlier = sm.OLS(calisto_ol_df[y], calisto_ol_df[x].assign(const=1))
o_outlier = sm.OLS(omega_ol_df[y], omega_ol_df[x].assign(const=1))

a_outlier_results = a_outlier.fit()
c_outlier_results = c_outlier.fit()
o_outlier_results = o_outlier.fit()


a_outlier_influence = OLSInfluence(a_outlier_results)
c_outlier_influence = OLSInfluence(c_outlier_results)
o_outlier_influence = OLSInfluence(o_outlier_results)

a_outlier_residuals = a_outlier_influence.resid_studentized_internal
c_outlier_residuals = c_outlier_influence.resid_studentized_internal
o_outlier_residuals = o_outlier_influence.resid_studentized_internal
```

```python
a_outlier_residuals
```

```python
#Combine all the residual from 3 dataframe
residual_outlier = a_outlier_residuals.append(c_outlier_residuals)
residual_outlier = residual_outlier.append(o_outlier_residuals)
residual_outlier = residual_outlier.sort_index()
residual_outlier
```

Calculate the IQR,lower limit and upper limit

```python
Q1 = residual_outlier.quantile(0.25)
Q3 = residual_outlier.quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit,upper_limit
```

Adding residuals column to the dataframe.

```python
outlier_df['residuals'] = residual_outlier
outlier_df
```

Identify which row is outliers by residual value must lower than lower limit or higher than upper limit.

```python
outlier_df[(outlier_df['residuals'] < lower_limit) | (outlier_df['residuals'] > upper_limit)]
```

## 5. Removing the outlier and saving the csv file

```python
no_outlier_df = outlier_df[(outlier_df['residuals'] > lower_limit) & (outlier_df['residuals'] < upper_limit)]
no_outlier_df
```

```python
no_outlier_df  = no_outlier_df.drop(['residuals'], axis=1)
no_outlier_df.shape
```

```python
no_outlier_df.to_csv('data_outliers_solution.csv', float_format='%.2f' ,encoding='utf-8', index=False)
```
